```python
import pandas as pd 

df = pd.read_csv("/Users/quentinnippert/Documents/mm_files/Telegram_analyse/STREAMLIT_APPLI/code_support/corpus_telegram_avril2026 - Sheet1.csv")

df
```

```python

from pathlib import Path

OUTPUT_FILE = Path("/Users/quentinnippert/Documents/mm_files/Telegram_analyse/STREAMLIT_APPLI/data/city_topics.csv")

df_work = df.copy()

# Чистим названия колонок
df_work.columns = df_work.columns.str.strip()

# Берём только скачанные espaces de communication
df_work = df_work[df_work["Téléchargé"] == "1"].copy()

df_work
```

```python

# Чистим город / регион
df_work["Ville/Région"] = df_work["Ville/Région"].astype(str).str.strip()

df_work = df_work[
    df_work["Ville/Région"].notna()
    & (df_work["Ville/Région"] != "")
    & (df_work["Ville/Région"].str.lower() != "nan")
].copy()

# Создаём city_id так же, как в cities.csv
df_work["city_id"] = (
    df_work["Ville/Région"]
    .str.lower()
    .str.replace(r"\s+", "_", regex=True)
    .str.replace(r"[^\w\-]", "", regex=True)
)

# Создаём новую таблицу city_topics.csv
city_topics = df_work[
    [
        "city_id",
        "Ville/Région",
        "Titre_support",
        "Type",
        "Type de communication",
        "Thématique",
        "Description",
    ]
].copy()

city_topics = city_topics.rename(
    columns={
        "Ville/Région": "city",
        "Titre_support": "titre_support",
        "Type": "type",
        "Type de communication": "type_communication",
        "Thématique": "thematique",
        "Description": "description",
    }
)

# На всякий случай чистим текстовые колонки
for col in city_topics.columns:
    city_topics[col] = city_topics[col].fillna("").astype(str).str.strip()

# Сохраняем
city_topics.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")

print(f"Ready: {OUTPUT_FILE}")
print(f"Rows saved: {len(city_topics)}")

city_topics.head()
```
