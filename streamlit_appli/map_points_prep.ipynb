```python
# 1. Копия датафрейма
df_work = df.copy()

# 2. Чистим колонку с локацией
df_work["Ville/Région"] = df_work["Ville/Région"].astype(str).str.strip()

# 3. Чистим колонку в таблице соответствий
cities_df["original_location"] = cities_df["original_location"].astype(str).str.strip()

# 4. Присоединяем city, city_id, lat, lon к основному df
df_work = df_work.merge(
    cities_df,
    left_on="Ville/Région",
    right_on="original_location",
    how="left"
)

# 5. Считаем количество строк по каждому городу
totals = (
    df_work
    .dropna(subset=["city_id"])
    .groupby(["city_id", "city", "lat", "lon"], as_index=False)
    .size()
    .rename(columns={"size": "total"})
)

# 6. Если хочешь получить итоговую таблицу для карты
cities_total_df = cities_df.merge(
    totals[["city_id", "total"]],
    on="city_id",
    how="left"
)

cities_total_df["total"] = cities_total_df["total"].fillna(0).astype(int)

# 7. Оставляем только нужные колонки
cities_total_df = cities_total_df[["city", "city_id", "lat", "lon", "total"]]

print(cities_total_df)
```

### Tableau 2. city_topics.csv

```python
# Копия основного датафрейма
df_work = df.copy()

# Чистим ключевые текстовые поля
df_work["Ville/Région"] = df_work["Ville/Région"].astype(str).str.strip()
df_work["Thématique"] = df_work["Thématique"].astype(str).str.strip()
df_work["Type"] = df_work["Type"].astype(str).str.strip()

cities_df["original_location"] = cities_df["original_location"].astype(str).str.strip()
cities_df["city_id"] = cities_df["city_id"].astype(str).str.strip()

# Подтягиваем city_id к основному df
df_work = df_work.merge(
    cities_df[["original_location", "city_id"]],
    left_on="Ville/Région",
    right_on="original_location",
    how="left"
)

# Убираем строки без города и без темы
df_topics = df_work.dropna(subset=["city_id"]).copy()
df_topics = df_topics[df_topics["Thématique"].notna()]
df_topics = df_topics[df_topics["Thématique"] != ""]
df_topics = df_topics[df_topics["Thématique"] != "nan"]

# Группировка по городу и тематике
city_topics_df = (
    df_topics
    .groupby(["city_id", "Thématique"], as_index=False)
    .agg(
        chat_count=("Type", lambda x: (x == "Chat").sum()),
        channel_count=("Type", lambda x: (x == "Chaine").sum()),
        bot_count=("Type", lambda x: (x == "Bot").sum()),
    )
    .rename(columns={"Thématique": "topic"})
    .sort_values(["city_id", "topic"])
    .reset_index(drop=True)
)

city_topics_df
```

```python
# Сохранение
city_topics_df.to_csv("city_topics.csv", index=False)
```
